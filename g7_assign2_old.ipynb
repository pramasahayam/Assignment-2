```python
# Import the wine dataset
import sklearn as the_machine # what is bro learning???
import numpy as np
import pandas as pd

wine_data = the_machine.datasets.load_wine()
features = wine_data.feature_names
wine_df = pd.DataFrame(data=wine_data.data, columns=features)
wine_df
```

```python
#shape of the dataset
shape = np.shape(wine_data.data)
print("Wine Data Shape: ", shape)

#size of data
size = np.size(wine_data.data)
print("Wine Data Size: ", size)

#dimensions of data
dim = np.ndim(wine_data.data)
print("Wine Data Dimensions: ", dim)
```

```python
#min, max, average for features and target labels
stats_df = pd.DataFrame(index = ['Minimum', 'Maximum', 'Average'])
for i in range(len(features)):
    #min, max, average of features
    name = features[i]
    stats_df[name] = [
        np.min(wine_df[name]),
        np.max(wine_df[name]),
        np.average(wine_df[name])
    ]

#min, max, average of target
stats_df['target'] = [
    np.min(wine_data.target),
    np.max(wine_data.target),
    np.average(wine_data.target)
]
stats_df
```

```python
#scatter matrix
plot = pd.plotting.scatter_matrix(wine_df[features], figsize=(30,30))
```

```python
#correlation
corr_matrix = wine_df.corr()
corr_matrix
```

```python
corr_disp = pd.DataFrame(index=['Correlation'])

for i in range(len(features)):
    name = features[i]
    for j in range(len(corr_matrix[name])):

        #elements will always have a 1 correlation with itself, filter out
        #filter out repeats since values across the matrix diagonal are the same
        if corr_matrix[name].iloc[j] == 1:
            break

        #check for strong correlation (x>0.7 or x<-0.7)
        if corr_matrix[name].iloc[j] > 0.7 or corr_matrix[name].iloc[j] < -0.7:
            corr_disp[f'{name} and {features[j]}'] = round(corr_matrix[name].iloc[j], 4)

corr_disp
```
